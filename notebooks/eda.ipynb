```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/heart_cleveland_upload.csv')
```

## Checking missing values

```python
df.isnull().sum()
```

## Checking target distribution

```python
sns.set()
df.condition.value_counts().plot(kind="bar", figsize=(5,5))
plt.title('Distribution of target variable',fontsize=15)
plt.xlabel('Heart Disease',fontsize=12)
plt.ylabel('Frequency',fontsize=12)
plt.xticks(rotation=0)
plt.show()
```

This means dataset is pretty balanced

## Rate of heart diseases to sex

```python
pd.crosstab(df.sex,df.condition).plot(kind="bar",figsize=(6,5))
plt.title('Heart Disease vs Sex',fontsize=15)
plt.xlabel('Sex',fontsize=12)
plt.xticks(rotation=0)
plt.legend(["No heart Disease", "Heart Disease"])
plt.ylabel('Frequency',fontsize=12)
plt.show()
```

There are much more men with heart diseases than women

## Heart diseases vs type of pain

```python
pd.crosstab(df.cp, df.condition).plot(kind="bar", figsize=(10, 5), ax=plt.gca())
plt.title("Heart Disease Frequency According To Chest Pain Experience",fontsize=15)
plt.legend(["Do not have heart disease", "Have heart disease"])
plt.xlabel("Chest Pain Experienced",fontsize=12)
plt.gca().set_xticklabels(
    labels=["Typical angina", "Non-anginal pain", "Atypical angina", "Asymptomatic"]
)
plt.xticks(rotation=0)
plt.ylabel("Frequency of Disease or Not",fontsize=12)
plt.show()
```

As we can see most of the pain which causes heart disease is asymptomatic

## Correlation map

```python
corr = df[['age','trestbps','chol','thalach','oldpeak']].corr()
plt.figure(figsize=(7,6))
sns.heatmap(corr, annot=True, vmin=-1.0, cmap="RdYlGn")
plt.show()
```

We can see that with higher age there is a bigger chance of heart attack, same with resting blood pressure and cholesterol. And if your maximum heart rate decreases it also might indicated to heart disease.

## Age vs heart rate

```python
plt.figure(figsize=(12,6))

plt.scatter(df.age[df.condition==1], df.thalach[df.condition==1],color='red')
plt.scatter(df.age[df.condition==0], df.thalach[df.condition==0],color='green')

plt.title('Age vs heart rate',fontsize=15)
plt.xlabel('Age',fontsize=12)
plt.ylabel('Max heart rate',fontsize=12)
plt.legend(['Disease','No Disease']);
```

This graph adds to correlation map and we can clearly see that with low max heart rate you probably have a heart disease.
